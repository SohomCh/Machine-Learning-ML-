```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df =pd.read_csv("homeprices.csv")
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price ")
plt.scatter(df.area,df.price,color='red',marker="+")
```

```python
price = df.price
reg = linear_model.LinearRegression()
reg.fit(df[['area']],price)
```

```python
reg.predict([[3300]])
```
